```python
# ---------------------------------------------------------
# Pattern Recognition Lab Assignment
# Histogram-Based Density Estimation + Bayes' Theorem
# ---------------------------------------------------------

# -----------------------------
# Section A Marks (50 students)
# -----------------------------
section_A = [
    8, 9, 10, 11, 12,
    10, 11, 12, 13, 14,
    11, 12, 13, 14, 15,
    12, 13, 14, 15, 16,
    13, 14, 15, 16, 17,
    14, 15, 16, 17, 18,
    15, 16, 17, 18, 19,
    16, 17, 18, 19, 20,
    17, 18, 19, 20, 21,
    18, 19, 20, 21, 22
]

# -----------------------------
# Section B Marks (50 students)
# -----------------------------
section_B = [
    15, 16, 17, 18, 19,
    17, 18, 19, 20, 21,
    18, 19, 20, 21, 22,
    19, 20, 21, 22, 23,
    20, 21, 22, 23, 24,
    21, 22, 23, 24, 25,
    22, 23, 24, 25, 26,
    23, 24, 25, 26, 27,
    24, 25, 26, 27, 28,
    25, 26, 27, 28, 29
]

# Prior Probabilities
P_A = 0.5
P_B = 0.5

# New student's mark
new_mark = 20

# ---------------------------------------------------------
# Histogram Bins
# (Same bins are used for both sections)
# ---------------------------------------------------------

bins = [
    (8, 10),
    (11, 13),
    (14, 16),
    (17, 19),
    (20, 22),
    (23, 25),
    (26, 29)
]


# ---------------------------------------------------------
# Function to calculate histogram frequencies
# ---------------------------------------------------------
def histogram_frequency(data, bins):

    frequencies = []

    for low, high in bins:

        count = 0

        for mark in data:
            if low <= mark <= high:
                count += 1

        frequencies.append(count)

    return frequencies


# ---------------------------------------------------------
# Function to find which bin contains the given mark
# ---------------------------------------------------------
def get_bin_index(value, bins):

    for i in range(len(bins)):

        low, high = bins[i]

        if low <= value <= high:
            return i

    return -1


# Histogram Frequencies
freq_A = histogram_frequency(section_A, bins)
freq_B = histogram_frequency(section_B, bins)

# ---------------------------------------------------------
# Display Histogram Frequencies
# ---------------------------------------------------------
print("========== HISTOGRAM FREQUENCIES ==========\n")

print("Section A")
for i in range(len(bins)):
    low, high = bins[i]
    print(f"{low:2d}-{high:2d} : {freq_A[i]}")

print()

print("Section B")
for i in range(len(bins)):
    low, high = bins[i]
    print(f"{low:2d}-{high:2d} : {freq_B[i]}")

# ---------------------------------------------------------
# Histogram-Based Likelihood
# ---------------------------------------------------------

bin_index = get_bin_index(new_mark, bins)

likelihood_A = freq_A[bin_index] / len(section_A)
likelihood_B = freq_B[bin_index] / len(section_B)

print("\n========== LIKELIHOOD ==========")
print(f"P(20 | A) = {likelihood_A:.4f}")
print(f"P(20 | B) = {likelihood_B:.4f}")

# ---------------------------------------------------------
# Bayes' Theorem
# ---------------------------------------------------------

evidence = (likelihood_A * P_A) + (likelihood_B * P_B)

posterior_A = (likelihood_A * P_A) / evidence
posterior_B = (likelihood_B * P_B) / evidence

print("\n========== POSTERIOR ==========")
print(f"P(A | 20) = {posterior_A:.4f}")
print(f"P(B | 20) = {posterior_B:.4f}")

# ---------------------------------------------------------
# Final Classification
# ---------------------------------------------------------

print("\n========== FINAL RESULT ==========")

if posterior_A > posterior_B:
    print("The student is more likely to belong to SECTION A.")
else:
    print("The student is more likely to belong to SECTION B.")
```
